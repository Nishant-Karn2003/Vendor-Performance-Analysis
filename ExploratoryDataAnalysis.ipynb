# Purchase Transactions

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mysql.connector

db=mysql.connector.connect( host="localhost",
                            username="root",
                            password="PASSWORD@629",
                            database="inventory")
cur = db.cursor()

query= """ select p.VendorNumber,p.VendorName,p.Brand,pr.Volume,p.PurchasePrice,pr.Price,sum(Quantity),sum(Dollars)
from inventory.purchase_prices as pr join inventory.purchases as p
on pr.Brand=p.Brand
where p.PurchasePrice>0

group by p.VendorNumber,p.Brand,p.VendorName,p.PurchasePrice, pr.Price,pr.Volume
order by sum(Dollars);"""
cur.execute(query)
data=cur.fetchall()
df=pd.DataFrame(data,columns=["VendorNumber","VendorName","Brand","Volume","PurchasePrice","ActualPrice","TotalQuantity","AmountSpent"])
display(df)


```

# Overall vendor performance

```python
import time
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mysql.connector

db=mysql.connector.connect( host="localhost",
                            username="root",
                            password="PASSWORD@629",
                            database="inventory")
cur = db.cursor()
start = time.time()
query= """ with pur as (select p.VendorNumber,p.VendorName,p.Brand,pr.Volume,p.PurchasePrice,pr.Price,sum(Quantity) as TotalQuantityPurchased,
sum(Dollars) as TotalAmountSpent
from inventory.purchase_prices as pr join inventory.purchases as p
on pr.Brand=p.Brand
where p.PurchasePrice>0
group by p.VendorNumber,p.Brand,p.VendorName,p.PurchasePrice, pr.Price,pr.Volume
order by sum(Dollars)),

sls as(select VendorNo,VendorName,Brand,SalesPrice,sum(ExciseTax)as TotalExcise
,sum(SalesQuantity) as TotalQuantitySold,sum(SalesDollars) as TotalEarned
from inventory.sales         
group by VendorNo,VendorName,Brand,SalesPrice
order by sum(SalesDollars)),

frg as(select VendorNumber,VendorName,round(sum(Freight),2) as Total_Freight from inventory.vendor_invoice
group by VendorNumber,VendorName
order by  VendorNumber)

select pur.VendorNumber,pur.VendorName,pur.Brand,pur.Price as ActualPrice,pur.PurchasePrice,
pur.TotalQuantityPurchased,pur.TotalAmountSpent,pur.Volume,
sls.SalesPrice as SalesPricePerQuantity,sls.TotalQuantitySold,sls.TotalEarned,
sls.TotalExcise,frg.Total_Freight
from
pur join sls
on pur.VendorNumber=sls.VendorNo
and pur.Brand=sls.Brand
join frg
on pur.VendorNumber=frg.VendorNumber
"""

cur.execute(query)
data=cur.fetchall()
vendor_sales_summary=pd.DataFrame(data,columns=["VendorNumber","VendorName","Brand","ActualPricePerProduct",
                              "PurPricePerProduct","TotalQuantityPurchased","TotalAmountSpent","Volume",
                             "SalesPricePerQuantity","TotalQuantitySold","TotalEarned","TotalExcise",
                             "TotalFreight"])
display(vendor_sales_summary)
end = time.time()
```

 <span style="font-size:90%">This query generates a vendor-wise sales and purchase summary,which is valuable for:</span>  
# Performance optimization:
- <span style="font-size:90%">This query involves heavy joins and aggregations on large datasets like sales and purchases</span>
- <span style="font-size:90%">Storing pre-aggregated results avoids repeated expensive computations</span>
- <span style="font-size:90%">Helps in analyzing,purchasesand pricing for different vendors and brands</span>
- <span style="font-size:90%">Future benefits for storing this data for faster dashboarding and reporting</span>
